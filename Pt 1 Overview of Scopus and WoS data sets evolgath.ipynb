```python
plt.subplot(2,2,1)
sns.histplot(data=scp_items, x="pubyear", y= 'author_count', color = 'darkorange')
plt.xlim(1990, 2022) # this seems not to be what we are looking for
plt.ylim(0,25)
plt.subplot(2,2,2)
sns.histplot(data=scp_items, x="pubyear", y= 'author_count', color = 'blueviolet')
plt.xlim(1990, 2022)
plt.ylim(0,25)

plt.subplot(2,2,3)
sns.distplot(scp_items['author_count'], hist=True, kde=False, color = 'darkorange',
            hist_kws={'edgecolor':'black'})
plt.subplot(2,2,4)
sns.distplot(wos_items['author_count'], hist=True, kde=False, color = 'blueviolet',
            hist_kws={'edgecolor':'black'})
plt.tight_layout()
```
