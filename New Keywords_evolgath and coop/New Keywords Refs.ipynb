Most references found by querying the abstracts of the citing papers. "evolution of cooperation" in abstracts results in the most references found.
In the following, we are considering only the unique item ids of the cited references.

```python
# Loop through dfs and count the number of unique cited ids
results = {}

for name, df in [('refs_emcoop', refs_emcoop),
                 ('refs_emcoop_ab', refs_emcoop_ab),
                 ('refs_emcoop_titles', refs_emcoop_titles),
                 ('refs_evolcoop', refs_evolcoop),
                 ('refs_evolcoop_ab', refs_evolcoop_ab),
                 ('refs_evolcoop_titles', refs_evolcoop_titles),
                 ('refs_evolgath', refs_evolgath),
                 ('refs_evolgath_ab', refs_evolgath_ab),
                 ('refs_evolgath_titles', refs_evolgath_titles)]:
    num_unique_cited_ids = df['item_id_cited'].nunique()
    results[name] = num_unique_cited_ids

# Convert the dictionary to a dataframe
df_results = pd.DataFrame.from_dict(results, orient='index', columns=['num_unique_cited_ids'])

# Display the dataframe
df_results
```

```python
for name, df in [('refs_emcoop', refs_emcoop),
                 ('refs_emcoop_ab', refs_emcoop_ab),
                 ('refs_emcoop_titles', refs_emcoop_titles),
                 ('refs_evolcoop', refs_evolcoop),
                 ('refs_evolcoop_ab', refs_evolcoop_ab),
                 ('refs_evolcoop_titles', refs_evolcoop_titles),
                 ('refs_evolgath', refs_evolgath),
                 ('refs_evolgath_ab', refs_evolgath_ab),
                 ('refs_evolgath_titles', refs_evolgath_titles)]:
    top_5 = df['item_id_cited'].value_counts().head(5)
    print(f'Top 5 cited item IDs for {name}:')
    print(top_5)
```
